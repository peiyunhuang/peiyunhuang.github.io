```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge( ride_df,city_df, how='left', on='city')
# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_area = combined_df[combined_df["type"] == "Urban"].groupby([combined_df["city"]])
urban_x_coordinate = urban_area["ride_id"].count()
urban_y_coordinate = urban_area["fare"].mean()
urban_bubbleplot_size = urban_area["driver_count"].mean()
#print(urban_area.head())

suburban_area = combined_df[combined_df["type"] == "Suburban"].groupby([combined_df["city"]])
suburban_x_coordinate = suburban_area["ride_id"].count()
suburban_y_coordinate = suburban_area["fare"].mean()
suburban_bubbleplot_size = suburban_area["driver_count"].mean()
#print(suburban_area.head())

rural_area = combined_df[combined_df["type"] == "Rural"].groupby([combined_df["city"]])
rural_x_coordinate = rural_area["ride_id"].count()
rural_y_coordinate = rural_area["fare"].mean()
rural_bubbleplot_size = rural_area["driver_count"].mean()
#print(rural_area.head())

# Build the scatter plots for each city types
plt.scatter(urban_x_coordinate, urban_y_coordinate, s=urban_bubbleplot_size * 5,
           label = "Urban", marker="o", facecolors="coral", edgecolors="black", alpha=0.75)
plt.scatter(suburban_x_coordinate, suburban_y_coordinate, s=suburban_bubbleplot_size * 5,
           label = "Suburban", marker="o", facecolors="lightskyblue", edgecolors="black", alpha=0.75)
plt.scatter(rural_x_coordinate, rural_y_coordinate, s=rural_bubbleplot_size * 5,
           label = "Rural", marker="o", facecolors="gold", edgecolors="black", alpha=0.75)

#Format the Scatterplot
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
legend = plt.legend(fontsize = 10, title= "City Types", loc= "best")
legend.legendHandles[0]._sizes = [75]
legend.legendHandles[1]._sizes = [75]
legend.legendHandles[2]._sizes = [75]
plt.text(43,36, "Note: Size of Bubble Corresponds to the Count of Drivers in a City", fontsize = 10)

#Save Scatterplot as a PNG
plt.savefig("PyberRideSharingData.png", bbox_inches="tight")
plt.show()
```

## Total Fares by City Type

```python
#Groupby with Type as the Index and Sum the Fares
type_ride_city = combined_df.groupby(["type"])
total_fare_city = type_ride_city["fare"].sum()


#Format the Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "coral"]
explode = (0, 0, 0.1)
plt.pie(total_fare_city, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=100)
plt.title("% of Total Fares by City Type")
#Save Pie Chart as PNG
plt.savefig("TotalFaresbyCityType")
plt.show()

```
